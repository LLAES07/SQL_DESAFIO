# **Importando librerias**

```python
import pandas as pd
import numpy as np
import sqlite3
```

# **GENERANDO LA BASE DE DATOS Y TABLAS**

```python
# Generando una database
connection = sqlite3.connect('foodie_fi.db')

# Comiteando la creaci√≥n
connection.commit()
```

```python
# Agregando el cursor para poder modificar y crear tablas
cursor = connection.cursor()
```

---

## **Creando y poblando tabla plans**

```python
# Consult para generar la tabla plans
cursor.execute('''

    CREATE TABLE plans (
               plan_id INTEGER,
               plan_name VARCHAR(13),
               price DECIMAL(5, 2)
               )

''')
connection.commit()
```

```python
# Lista de tuplas para poblar la tabla
plans_data = [
    ('0', 'trial', '0'),
    ('1', 'basic monthly', '9.90'),
    ('2', 'pro monthly', '19.90'),
    ('3', 'pro annual', '199'),
    ('4', 'churn', np.nan)
]


# Usando el cursor con executemany para poder poblar en una sola linea los datos de plans_data
cursor.executemany('INSERT INTO plans VALUES ( ?, ?, ?)', plans_data)


# Generando los cambios
connection.commit()
```

```python


def verifica_tabla(table_name: str):
    """
    Funcion que devuelve los datos de una tabla en una base de datos limitada por 5 registros
    """

    df = pd.read_sql_query(f'''SELECT * FROM {table_name} LIMIT 5''', connection)
    return df
```

```python
# Revisando la tabla generada
verifica_tabla('plans')
```

# **2. ¬øCu√°l es la distribuci√≥n mensual de los valores de start_date del plan de prueba en nuestro conjunto de datos? Usa el inicio del mes como valor para agrupar.**

```python
# Distribucion mensual = planes de prueba por mes


query = '''

    SELECT
        strftime('%Y-%m-01', start_date) AS inicio_mes,
        COUNT(DISTINCT customer_id) AS total

     FROM
         subscriptions

    WHERE plan_id = (SELECT plan_id FROM plans WHERE plan_name='trial')
    GROUP BY inicio_mes

'''

pd.read_sql_query(query, connection)
```

### ***RESPUESTA PANDAS***

```python
# Aseguramos que start_date este en formato fecha
df_subs['start_date'] = pd.to_datetime(df_subs['start_date'])

# Hacemos un merge entre las dos tablas
df_merge = pd.merge(df_subs, df_plans, on='plan_id')

# Filtramos el merge por aquellos planes donde sea trial
df_merge = df_merge.loc[df_merge['plan_name']=='trial', :]

# Agrupamos de forma mensual y utilizamos size para conocer la cantidad de clientes.
df_merge.groupby(df_merge['start_date'].dt.strftime('%Y-%m-01')).size().rename('total_clientes').to_frame()
```

üöÄ **Explicaci√≥n**

Necesitamos conocer la distribuci√≥n mensual de nuestros clientes con el plan trial, para esto necestimaos primero filtrar los datos por aquellos clientes que tengan el plan name trial, una vez realizado esto agrupamos por la fecha y contamos la cantidad de clientes unicas por mes

## **3. ¬øQu√© planes ocurren despu√©s del a√±o 2020 en nuestro conjunto de datos? Muestra el desglose por conteo de eventos para cada nombre de plan**

### ***RESPUESTA SQL***

```python

query = '''

    SELECT
        p.plan_name,
        COUNT(*) AS total_planes
     FROM
        subscriptions s
    INNER JOIN 
        plans p USING(plan_id)

    WHERE strftime('%Y', start_date)>'2020'
    GROUP BY p.plan_id, p.plan_name

'''

pd.read_sql_query(query, connection)
```

### ***RESPUESTA PANDAS***

```python
# Merge de las tablas
df_merge = pd.merge(df_subs, df_plans, on='plan_id')

# Filtrado por fechas superiores a 2020
df_merge = df_merge.loc[df_merge['start_date'].dt.year>2020, :]

# Agurpando por plan name y contando los registros  
df_merge.groupby(df_merge['plan_name']).size().rename('total_planes').to_frame().reset_index()
```

üöÄ **Explicaci√≥n**

Nos preguntan por los planes que se dan despu√©s del a√±o 2020 y su cantidad respectiva. Para esto primero realizamos un join entre las tablas posterior a este join lo filtramos por fechas superiores a 2020, con esto ya podemos agrupar por el nombre del plan y contar los registros.

## **4. ¬øC√∫al es la cantidad y el porcentaje de de  clientes que  desertado, redondeado a 1 decimal ?**

### ***RESPUESTA SQL***

```python
query = '''
    
    SELECT
        
        COUNT(DISTINCT s.customer_id) AS total_clientes_desertores,
        ROUND(COUNT(*)*100/ (SELECT COUNT(DISTINCT customer_id) FROM subscriptions ), 1) AS porcentaje
     FROM
        subscriptions s
    INNER JOIN 
        plans p USING(plan_id)
    WHERE 
        p.plan_name ='churn'


'''

pd.read_sql_query(query, connection)
```

### ***RESPUESTA PANDAS***

# **10. ¬øPuedes desglosar a√∫n m√°s este valor promedio en per√≠odos de 30 d√≠as (es decir, de 0 a 30 d√≠as, de 31 a 60 d√≠as, etc.)?**

## ***RESPUESTA SQL***

```python
query = '''

    WITH pro_anual_costumers AS (
    SELECT
        *
     FROM
        subscriptions s
    INNER JOIN 
        plans p USING(plan_id)
    WHERE 
        plan_name = 'pro annual'),

    pro_customers_first_date AS (
    SELECT
        *
     FROM
        subscriptions s
    INNER JOIN 
        plans p USING(plan_id)
    WHERE 
        plan_name = 'trial' AND s.customer_id IN ( SELECT DISTINCT customer_id FROM  pro_anual_costumers) ),
    
    CT1 AS (

    SELECT
        *,
        julianday(p.start_date) - julianday(f.start_date) as days,
        CASE
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 0 AND 30 THEN '0-30'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 31 AND 60 THEN '30-60'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 61 AND 90 THEN '60-90'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 91 AND 120 THEN '90-120'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 121 AND 150 THEN '120-150'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 151 AND 180 THEN '150-180'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 181 AND 210 THEN '180-210'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 211 AND 240 THEN '210-240'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 241 AND 270 THEN '240-270'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 271 AND 300 THEN '270-300'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 301 AND 330 THEN '300-330'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) BETWEEN 331 AND 360 THEN '330-360'
            WHEN (julianday(p.start_date) - julianday(f.start_date)) > 360  THEN '360 +'

            ELSE 'fuera_rango'
            END AS rangos
    FROM pro_anual_costumers p
    LEFT JOIN  pro_customers_first_date f
     ON p.customer_id = f.customer_id )


    SELECT
        rangos, 
        COUNT(days) AS valores_rangos

    FROM
        CT1
    GROUP BY 
        rangos


        

'''

pd.read_sql_query(query, connection)
```

## ***RESPUESTA PANDAS***

```python
df_merge = pd.merge(df_subs, df_plans, on='plan_id')


clientes_pro_anual = df_merge[df_merge['plan_name'] == 'pro annual']
clientes_trial = df_merge[
    (df_merge['plan_name'] == 'trial') &
    (df_merge['customer_id'].isin(clientes_pro_anual['customer_id'].unique()))
]

pro_anual_customers = clientes_trial.rename(columns={'start_date': 'start_date_pro'})
trial_customers = clientes_trial.rename(columns={'start_date': 'start_date_trial'})

ct1 = pd.merge(
    pro_anual_customers,
    trial_customers[['customer_id', 'start_date_trial']],
    on='customer_id',
    how='left'
)

ct1['start_date_pro'] = pd.to_datetime(ct1['start_date_pro'])
ct1['start_date_trial'] = pd.to_datetime(ct1['start_date_trial'])

ct1['days'] = (ct1['start_date_pro'] - ct1['start_date_trial']).dt.days


def clasificar_rango(d):
    if pd.isna(d): return 'fuera_rango'
    elif 0 <= d <= 30: return '0-30'
    elif 31 <= d <= 60: return '30-60'
    elif 61 <= d <= 90: return '60-90'
    elif 91 <= d <= 120: return '90-120'
    elif 121 <= d <= 150: return '120-150'
    elif 151 <= d <= 180: return '150-180'
    elif 181 <= d <= 210: return '180-210'
    elif 211 <= d <= 240: return '210-240'
    elif 241 <= d <= 270: return '240-270'
    elif 271 <= d <= 300: return '270-300'
    elif 301 <= d <= 330: return '300-330'
    elif 331 <= d <= 360: return '330-360'
    elif d > 360: return '360 +'
    else: return 'fuera_rango'

ct1['rangos'] = ct1['days'].apply(clasificar_rango)

result = ct1.groupby('rangos')['days'].count().reset_index(name='valores_rangos')
result
```
